```python
# Import the libraries
import pandas as pd
import selenium
import time
from selenium import webdriver

import warnings
warnings.filterwarnings("ignore")
```

```python
# Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\saman\Downloads\chromedriver_win32\chromedriver.exe") 
```

```python
# Load the url
driver.get("https://www.yatra.com/flight-schedule/")
```

```python
# exploring the locations
all_loc_flight=[]

for i in driver.find_elements_by_xpath("//li[@class='mb10']/a")[0:]:
    all_loc_flight.append(i.get_attribute('href'))
```

```python
airline_name = []
date_of_journey = []
source = []
destination = []
departure_time = []
arrival_time = []
duration = []
total_stops = []
price = []

for url in all_loc_flight:
    driver.get(url)

    try:
        for i in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']"):
            airline_name.append(i.text)
                                                     
        for i in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p"):
            source.append(i.text)                                        
  
        for i in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]"):
            destination.append(i.text)                                      
 
        for i in driver.find_elements_by_xpath("//div[@class='i-b pr']"):
            departure_time.append(i.text)                                      
  
        for i in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']"):
            arrival_time.append(i.text.split('\n')[0])                                        
  
        for i in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']"):
            duration.append(i.text)                                         

        for i in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span"):
            total_stops.append(i.text.split(" ")[0])                                     
 
        for i in driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']"):
            price.append(i.text.replace(",",""))                                      
            
        details=driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")
        date=driver.find_elements_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']/p")[0].text.split(",")[1][1:]
            
        for i in range(len(details)):
            date_of_journey.append(date)
    except:
        continue
```

```python
links=["https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&preferred=&destinationCountry=IN&flight_depart_date=24%2F04%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
       "https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=GOI&preferred=&destinationCountry=IN&flight_depart_date=04%2F06%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
       "https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BLR&preferred=&destinationCountry=IN&flight_depart_date=08%2F06%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
       "https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=MAA&preferred=&destinationCountry=IN&flight_depart_date=14%2F06%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
       "https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=PNQ&originCountry=IN&destination=CCU&preferred=&destinationCountry=IN&flight_depart_date=20%2F06%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
       "https://flight.yatra.com/air-search-ui/seodom/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=PAT&originCountry=IN&destination=BLR&preferred=&destinationCountry=IN&flight_depart_date=20%2F06%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=seo",
      ]

count=0

for lk in links:
    try:
        driver.get(lk)
    except:
        driver.get(lk)
    time.sleep(10)
    for i in range(30):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)

    details=driver.find_elements_by_xpath("//div[@class='flightItem border-shadow pr']")
    date=driver.find_elements_by_xpath("//div[@class='day-li text-center cursor-pointer pr active font-primary-color']/p")[0].text.split(",")[1][1:]
    
    airline_name.extend([w.text for w in driver.find_elements_by_xpath("//span[@class='i-b text ellipsis']")])
    departure_time.extend([w.text for w in driver.find_elements_by_xpath("//div[@class='i-b pr']")])
    arrival_time.extend([w.text.split("\n")[0] for w in driver.find_elements_by_xpath("//p[@class='bold fs-15 mb-2 pr time']")])
    source.extend([w.text for w in driver.find_elements_by_xpath("//div[@class='i-b col-4 no-wrap text-right dtime col-3']/p")])
    destination.extend([w.text for w in driver.find_elements_by_xpath("//div[@class='i-b pdd-0 text-left atime col-5']/p[2]")])
    duration.extend([w.text for w in driver.find_elements_by_xpath("//p[@class='fs-12 bold du mb-2']")])
    price.extend([w.text.replace(",","") for w in driver.find_elements_by_xpath("//div[@class='i-b tipsy fare-summary-tooltip fs-18']")])
    total_stops.extend([ w.text.split(" ")[0] for w in driver.find_elements_by_xpath("//div[@class=' font-lightgrey fs-10 tipsy i-b fs-10']/span")])
        
    for i in range(len(details)):
        date_of_journey.append(date)
    count+=1
    print("Count:",count)
```

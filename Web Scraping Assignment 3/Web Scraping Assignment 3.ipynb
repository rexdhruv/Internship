Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”.
```python
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("http://www.amazon.in")
time.sleep(2)

search_bar= driver.find_element_by_id("twotabsearchtextbox")
search_bar.send_keys("guitar")
search_btn=driver.find_element_by_id("nav-search-submit-button")
search_btn.click()
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='a-link-normal s-link-style a-text-normal']"):
    urls.append(i.get_attribute('href'))
print("Length:", len(urls))
print(urls[:2])

time.sleep(2)
brands=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
brand=[]
for i in brands:
    if i.text is None :
        brand.append("-") 
    else:
        brand.append(i.text)
    
print(" Length:", len(brand))
print(brand[0:2])

time.sleep(2)
cost=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
price=[]
for i in cost:
    if i.text is None :
        price.append("-") 
    else:
        price.append(i.text)
    
print('\n',"Length:", len(price))
print(price[0:2])

time.sleep(2)
exp_delevery=driver.find_elements_by_xpath("//span[@class='a-text-bold']")
exp_del=[]
for i in exp_delevery:
    if i.text is None:
        exp_del.append("-")
    else:
        exp_del.append(i.text)
        
print('\n','Length:', len(exp_del))
print(exp_del[0:2])
```

## 4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
url = 'https://www.flipkart.com'
driver.get(url)
time.sleep(2)

close_popup = driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')
close_popup.click()

search_bar=driver.find_element_by_name("q")
search_bar.send_keys("mobile phones")
search_btn=driver.find_element_by_class_name("_34RNph")
search_btn.click()

time.sleep(20)
soup = BeautifulSoup(driver.page_source, 'html.parser')

time.sleep(20)
results = soup.find_all('a',{'class':"_1fQZEK"})
len(results)

time.sleep(20)
item = results[0]
```

```python
def extract_phone_model_info(item):
    
    model = item.find('div',{'class':"_4rR01T"}).text

    ram = item.find('li',{'class':"rgWa7D"}).text[0:item.find('li',{'class':"rgWa7D"}).text.find('|')]

    storage = item.find('li',{'class':"rgWa7D"}).text[item.find('li',{'class':"rgWa7D"}).text.find('|')+1:][0:10].strip()

    expandable = item.find('li',{'class':"rgWa7D"}).text[item.find('li',{'class':"rgWa7D"}).text.find('|')+1:][13:]

    display = item.find_all('li')[1].text.strip()

    camera = item.find_all('li')[2].text.strip()

    battery = item.find_all('li')[3].text

    processor = item.find_all('li')[4].text.strip()

    warranty = item.find_all('li')[-1].text.strip()

    price = item.find('div',{'class':'_30jeq3 _1_WHN1'}).text
    
    links_with_text = [a['href'] for a in soup.find_all('a', href=True) if a.text]

    result = (model,ram,storage,expandable,display,camera,battery,processor,warranty,price,links_with_text)
    return result

records_list = []
results = soup.find_all('a',{'class':"_1fQZEK"})
for item in results:
    records_list.append(extract_phone_model_info(item))
```

```python
df=pd.DataFrame(records_list,columns=['model','ram',"storage","expandable","display","camera","battery","processor","warranty","price","product_url"])
df.index=df.index+1
df
```

```python
# saving the dataframe
df.to_csv('smartphone.csv')
```

```python
# Close the Web Driver
driver.close()
```

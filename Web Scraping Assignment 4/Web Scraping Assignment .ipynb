#### 2. Scrape the details team India’s international fixtures from bcci.tv. Url = https://www.bcci.tv/. You need to find following details:
    A) Match title (I.e. 1st ODI)
    B) Series
    C) Place
    D) Date
    E) Time
    Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
import time
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("https://www.bcci.tv/")

international = driver.find_element_by_xpath('//li[@class="nav-item ml-0"]')
international.click()
time.sleep(5)

title=[]
series=[]
place=[]
date=[]
time=[]

Series=driver.find_elements_by_xpath('//span[@class="ng-binding"]')
for i in Series:
    series.append(i.text)

Place=driver.find_elements_by_xpath('//div[@class="fix-place ng-binding ng-scope"]')
for i in Place:
    title.append(i.text.split("-",2)[0])
    place.append(i.text.split(" ",3)[3])
    
Date=driver.find_elements_by_xpath('//h5[@class="ng-binding"]')
for i in Date:
    date.append(i.text)
    
Time=driver.find_elements_by_xpath('//h5[@class="text-right ng-binding"]')
for i in Time:
    time.append(i.text)

len(series),len(title),len(place),len(date),len(time)
```

```python
driver.close()

df = pd.DataFrame({"Title":title,
                  "Series":series,
                  "Place":place,
                  "Date":date,
                  "Time":time})
df.index=df.index+1
df
```

# 7. Scrape the details of Data science recruiters from naukri.com. Url = https://www.naukri.com/ 
You have to find the following details: A) Name B) Designation C) Company D) Skills they hire for E) Location 
Note: - From naukri.com homepage click on the recruiters option and the on the search pane type Data science and click on search. All this should be done through code

```python
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
driver.get('https://www.naukri.com/mnjuser/homepage')
```

```python
recruiters=driver.find_element_by_xpath("//a[@title='Search Recruiters']")
driver.get(recruiters.get_attribute('href'))

search=driver.find_element_by_xpath("//input[@class='sugInp']")
search.send_keys("Data Science")
time.sleep(2)

button=driver.find_element_by_xpath("//button[@id='qsbFormBtn']")
button.click()
time.sleep(2)
```

```python
name=[]
try:
    names=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[1]")
    for i in names:
        name.append(i.text)
except NoSuchElementException:          
    name.append('-')
time.sleep(2)

designation=[]
try:
    designation_list=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/span[1]")
    for i in designation_list:
        designation.append(i.text)
except NoSuchElementException:     
    designation.append('-')
time.sleep(2)

company=[]
try:
    companies=driver.find_elements_by_xpath("//div[@class='recInfo']/div/p/a[2]")
    for i in companies:
        company.append(i.text)
except NoSuchElementException:
    company.append('-')
time.sleep(2)

location=[]
soup=BeautifulSoup(driver.page_source,'html.parser')
l=soup.find_all('div',attrs={'vcard'})
for i in l:
    try:
        location.append(i.find('small').text)
    except AttributeError :
        location.append("-")
time.sleep(2)

skill=[]
try:
    skills=driver.find_elements_by_xpath("//div[@class='recInfo']/div[2]")
    for i in skills:
        skill.append(i.text)
except NoSuchElementException:      
    skill.append('-')
```

```python
driver.close()
df=pd.DataFrame({'Name':name,
                'Designation':designation,
                'Company':company,
                'Skills they hire':skill,
                'Location':location,
                })
df.index=df.index+1
df
```

# 9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/

You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
name=[]
years=[]
genre=[]
run_time=[]
rating=[]
votes=[]

try:
    names=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
    for i in names:
        name.append(i.text)
except NosuchElementException:
    name.append('-')
time.sleep(2)

try:
    year=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")
    for i in year:
        years.append(i.text)
except NosuchElementException:
    years.append('-')
time.sleep(2)

try:
    genres=driver.find_elements_by_xpath("//span[@class='genre']")
    for i in genres:
        genre.append(i.text)
except NoSuchElementException:
    genre.append('-')
time.sleep(2)

try:
    run=driver.find_elements_by_xpath("//span[@class='runtime']")
    for i in run:
        run_time.append(i.text)
except NoSuchElementException:
    run_time.append('-')
time.sleep(2)

try:
    ratings=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
    for i in ratings:
        rating.append(i.text)
except NoSuchElementException:
    rating.append('-')
time.sleep(2)

try:
    vote=driver.find_elements_by_xpath("//span[@name='nv']")
    for i in vote:
        votes.append(i.text)
except NoSuchElementException:
    votes.append('-')
```

```python
driver.close()
df=pd.DataFrame({"Name":name,
                "Year Span":years,
                "Genre":genre,
                "Run Time":run_time,
                "Ratings":rating,
                "Votes":votes})
df.index=df.index+1
df
```

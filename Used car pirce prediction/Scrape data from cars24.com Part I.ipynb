```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re

# importing warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r"C:\Users\dell\Downloads\chromedriver_win32\chromedriver.exe")
driver.get("https://www.cars24.com/")
time.sleep(2)

cross_button=driver.find_element_by_xpath('//div[@class="float-right TBxAN"]')
cross_button.click()
```

```python
# exploring the locations
all_cars_loc_data=[]

for i in driver.find_elements_by_xpath("//ul[@class='_2s4G4']/li/a")[10::11]:
    all_cars_loc_data.append(i.get_attribute('href'))
```

```python
all_cars_loc_data
```

```python
location=[]
brand_n_model=[]
km_owner_fuel=[]
variant_mannual_auto=[]
price=[]
```

```python
try:
    for details in all_cars_loc_data:
        driver.get(details)
        time.sleep(5)
        loc= driver.find_element_by_tag_name("h1").text

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(60)
        for _ in range(1000):
            driver.execute_script("window.scrollBy(0,10000)")
        
       
        for i in range(len(driver.find_elements_by_xpath("//h2[@class='_3FpCg']"))):
            location.append(loc)
        
        try:
            for i in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
                brand_n_model.append(i.text)
                                                     
        except:
            brand_n_model.append("-")
            print("Error1")
        
        try:
            for i in driver.find_elements_by_xpath("//ul[@class='bVR0c']"):
                km_owner_fuel.append(i.text)
        except:
            km_owner_fuel.append("-")
            print("Error2")
        
        try:
            for i in driver.find_elements_by_xpath("//p[@class='cvakB']"):
                variant_mannual_auto.append(i.text)
        except:
            variant_mannual_auto.append("-")
            print("Error3")

        for i in driver.find_elements_by_xpath("//div[@class='_7udZZ']/span"):
            price.append(i.text)        
        
except:
    print("Error Occured!!!") 
```

```python
len(location), len(brand_n_model), len(km_owner_fuel), len(variant_mannual_auto), len(price)
```

```python
manufacture_year=[]
car_model=[]
brand=[]
km_driven=[]
no_of_owner=[]
fuel_type=[]
variant=[]
manual_auto=[]
Location=[]
nos=[]
Car_Price=[]

for i in brand_n_model:
    car_model.append(i.split(" ")[1])
    manufacture_year.append(int(i.split(" ")[0]))
    brand.append(i.split(" ")[2])
for i in km_owner_fuel:
    km_driven.append(int(i.split("\n")[0].split(" ")[0].replace(",","")))
    no_of_owner.append(i.split("\n")[1])
    fuel_type.append(i.split("\n")[2])
for i in variant_mannual_auto:
    check=i.split(" ")[-1]
    if (check=='Manual' or check=='Automatic'):
        manual_auto.append(check)
        variant.append(i.split(" ")[:-1])
    elif i.split(" ")[-1]=="AT":
        manual_auto.append("Automatic")
        variant.append(i.split(" "))
    else:
        manual_auto.append("Manual")
        variant.append(i.split(" "))
for i in location:
    s=""
    l=i.split(" ")[3:]
    for _ in range(len(l)):
        s+=l[_]+" "
    Location.append(s)
for i in no_of_owner:
    nos.append(int(i[0]))
for i in price:
    Car_Price.append(int(i[1:].replace(",","")))
```

```python
v=[]
for i in variant:
    k=""
    for j in range(len(i)):
        k+=i[j]
        k+=" "
    v.append(k)
```

```python
# creating the dataframe
df=pd.DataFrame()
df['Brand']=car_model
df['model']=brand
df['variant']=v
df['Transmission']=manual_auto
df['Year of Manufacturing']=manufacture_year
df['Driven_in_km']=km_driven
df['Fuel_type']=fuel_type
df['Number of owner']=nos
df['Location']=Location
df['Price']=Car_Price
df
```

```python
df.to_csv('Cars24.csv')
```
